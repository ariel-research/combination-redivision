```python
import pandas
df = pandas.read_csv('rishon-letzion.csv')
print("Raw table: ", df)
```

```python
group_by_agent_id = df.groupby("in_agent_id")[["in_agent_id"]].count().rename(columns={"in_agent_id": "item_count"})
agent_ids = list(group_by_agent_id.index)
n = len(agent_ids)
print(f"n={n} agents: {agent_ids}")
print(group_by_agent_id)
```

```python
missing_agent_ids = set(range(455)) - set(agent_ids)
print(sorted(missing_agent_ids))
```

```python
group_by_item_id = df.groupby("out_lot_index")[["out_lot_index"]].count().rename(columns={"out_lot_index": "agent_count"})
item_ids = list(group_by_item_id.index)
m = len(item_ids)
print(f"m={m} items: {item_ids}")
print(group_by_item_id)
```

```python
item_agent_pairs = group_by_item_id["agent_count"].sum()
agent_item_pairs = group_by_agent_id["item_count"].sum()
print(f"item_agent_pairs={item_agent_pairs}, agent_item_pairs={agent_item_pairs}")
```

```python
group_by_items_per_agent = group_by_agent_id.groupby("item_count")[["item_count"]].count().rename(columns={"item_count": "agent_count"})
group_by_items_per_agent
```

```python
group_by_agents_per_item = group_by_item_id.groupby("agent_count")[["agent_count"]].count().rename(columns={"agent_count": "item_count"})
group_by_agents_per_item
```

```python
map_item_to_max_value = df.groupby("out_lot_index")["out_total_value_of_lot"].max().to_dict() # .rename(columns={"out_lot_index": "agent_count"})
map_item_to_min_value = df.groupby("out_lot_index")["out_total_value_of_lot"].min().to_dict() # .rename(columns={"out_lot_index": "agent_count"})
print(map_item_to_max_value)
print(map_item_to_min_value) # They are identical

map_item_to_value = {item: int(v.replace(",","")) for item,v in map_item_to_max_value.items()}
print(map_item_to_value)
```

```python
map_agent_to_entitlement = df.groupby("in_agent_id")["in_relative_ownership"].max().to_dict()
map_agent_to_entitlement
```

```python
total_value = sum(map_item_to_value.values())
print(f"total_value={total_value}")
```

```python
agents = [(agent,map_agent_to_entitlement[agent]) for agent in agent_ids]
items  = [(item,1) for item in item_ids]
iteration = 1
from collections import defaultdict, Counter
map_agent_to_num_of_items = defaultdict(int)
while agents and items:
    agent,entitlement = agents[0]
    due_value = entitlement*total_value
    item,fraction = items[0]
    fraction_value = fraction*map_item_to_value[item]
    print(f"Agent {agent} deserves {entitlement} of {total_value}, which is worth {due_value}")
    if fraction_value==due_value:
        print(f"{iteration}. Agent {agent} gets {fraction} of {item}, which is worth {fraction_value}")
        map_agent_to_num_of_items[agent] += 1
        agents.pop(0)
        items.pop(0)
    elif fraction_value>due_value:
        fraction_to_give = due_value/map_item_to_value[item]
        print(f"{iteration}. Agent {agent} gets {fraction_to_give} of {item}, which is worth {due_value}")
        map_agent_to_num_of_items[agent] += 1
        agents.pop(0)
        items[0] = (item,fraction - fraction_to_give)
    else:  # due_value > fraction_value
        print(f"{iteration}. Agent {agent} gets {fraction} of {item}, which is worth {fraction_value}")
        map_agent_to_num_of_items[agent] += 1
        entitlement_satisfied = fraction_value/total_value
        items.pop(0)
        agents[0] = (agent, entitlement-entitlement_satisfied)
    iteration += 1

print(map_agent_to_num_of_items)
print(Counter(map_agent_to_num_of_items.values()))
```
